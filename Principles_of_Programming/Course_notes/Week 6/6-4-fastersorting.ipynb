### Quick Sort
Can you imagine a better name for an algorithm? Quick sort is faster than any of the sorts we've seen so far.

Here is the idea: start by choosing a *pivot* item – for the purposes of the algorithm, any item will do, let's suppose we pick the last item in the list:

<table><tbody><tr><td>37</td><td>42</td><td>9</td><td>19</td><td>35</td><td>4</td><td>53</td><td><b><u>22</u></b></td></tr></tbody></table>

Next, write the pivot item down on its own:

<table><tbody><tr><td><b><u>22</u></b></td></tr></tbody></table>

Now we need to go through each item in the original list, and write it on the correct *side* of the pivot. We aren't comparing the items to each other, only to the pivot. So we get:

<table><tbody><tr><td>9</td><td>19</td><td>4</td><td><b><u>22</u></b></td><td>35</td><td>37</td><td>53</td><td>42</td></tr></tbody></table>

This is one iteration – the list has been split into two unsorted sublists. Now we can apply the same process again to both halves. Hopefully, this is screaming *recursion* to you! We want to *quick sort* the left list, then quick sort the right list. This is one situation where implementing this with a loop will actually be significantly more tricky.

#### Partitioning
So how is this implemented? Hopefully the recursion hasn't scared you off. But even then, the details are lacking on how to actually sort the items to the left and right of the pivot. It's easy to do this on paper with a few numbers, how do we do it with a limited amount of memory – ideally in-place?

There are actually various ways to do this, and the one I'm going to use here is called Lomuto partitioning. It can handle any pivot, but the simplest version works when it is at one of the ends, so using the last element here is useful for simplicity.

The basic idea is to create a “pointer” which starts at the first position of the list – in other words, a variable set to `0`. Then we go through each item in the list in turn and compare its value with the *pivot*. If the value is smaller than the pivot, then we swap the value with the value in the *pointer* position, *and* we increase the pointer by one. Eventually we reach the last item (the pivot), at which point we swap it with the final position of the pointer. By doing this we will have ensured that all the items smaller than the pivot end up to the left of the pivot in the list. Additionally, the pivot is guaranteed to be in its final position, so we can sort the sublists left and right of the pivot without including it.

#### Implementation
And here is a Python implementation. Notice the implementation details again: much like the binary sort, we are able to avoid creating new lists by using `start` and `end` variables which tell us which section of the list to sort.

# Note:

The function **sorts in-place**, meaning it **directly modifies the list that was passed i**n**, rather than creating and returning a new sorted list. So the **caller never needs to capture a return value**

````python
def quick_sort(my_list, start=None, end=None):
    if start is None:
        start = 0
    if end is None:
        end = len(my_list)
        
    if end-start <= 1:
        return
        
    ptr = start
    pivot = end-1
    for i in range(start, end):
        if my_list[i] < my_list[pivot]:
            my_list[i], my_list[ptr] = my_list[ptr], my_list[i]
            ptr += 1
            
    my_list[pivot], my_list[ptr] = my_list[ptr], my_list[pivot]
    
    quick_sort(my_list, start=start, end=ptr)
    quick_sort(my_list, start=ptr+1, end=end)
    
my_list = [37, 42, 9, 19, 35, 4, 53, 22]
quick_sort(my_list)
print(my_list)
````

### Quick Sort Complexity
If the name weren't enough, we've already spoiled the fact that quick sort has a better complexity class than selection sort and insertion sort. But can you work out exactly what it is?

***Try to work out the complexity of quick sort before continuing.***

Quick sort is efficient because it takes a *divide and conquer* approach: it finds a way to split the list into two halves, then treat each of those the same – much like binary search. This is generally a good approach to take to find a fast algorithm, if you can.

As for the exact complexity, in the previous $O(n^2)$ algorithms we ended up noticing a pattern of nested for loops, each one contributing $n$ to the overall picture. Again the body of the quick sort algorithm includes another for loop which spans the length of the list, so it contributes at least $O(n)$.

But the “outer” loop is obscured slightly because of the recursion. The way to think about this is to think back to binary search again. Provided quick sort manages to split the list roughly in half each time it repeats, the total number of halving operations will be *logarithmic*. Of course, unlike binary search, we check both halves – but in big O notation, $O(2 \log(n))$ is the same as $O(\log(n))$

So each run of the function performs $n$ comparisons, and we do this $\log(n)$ times, giving us the final average and best case complexity: $O(n \log(n))$. 

We mentioned before that $O(\log(n))$ is significantly better than $O(n)$, and likewise, $O(n \log(n))$ is significantly better than $O(n^2)$. In fact, it has been proven that this is the best possible complexity for a sorting algorithm in the average case – it is not possible to create an algorithm which sorts in $O(n)$ time for example.

There are a couple of sticking points left to deal with for quick sort. First of all, we mentioned that $O(n \log(n))$ was the best and average case complexity. In the worst case, the pivot is either the largest or smallest item in the list. If this happens on every iteration, then the complexity degrades to $O(n^2)$! Unfortunately for our method of choosing the pivot, if the list is already sorted or in reverse order we will be guaranteed to see this behaviour. This is a good reason to pick a different pivot, and there is lots of work on trying to determine the best strategy here.

Finally, does quick sort use any additional memory? You might be tempted to say no, it *looks* like an in-place algorithm. But actually, these recursive function calls use memory! Every time the function calls itself, Python has to remember the local variables and which line of execution to return to. Because the function calls itself a logarithmic number of times based on the size of the list, quick sort in this implementation actually uses $O(\log(n))$ additional memory. 

This shouldn't be too much of a concern – again $\log(n)$ is really small compared to $n$, especially for large values of $n$, which would normally be our biggest worry. An iterative implementation of the algorithm would not be able to escape this requirement either – we would need to use something like a stack to keep track of the values ourselves (which is exactly how the recursion works, anyway).

### Pivot selection strategies — this is where it gets interesting:
- First or last element — simple but terrible on already-sorted input, which hits that O(n²) worst case
- Random pivot — swap a random element to the end before partitioning, makes worst case extremely unlikely in practice
- Median-of-three — look at first, middle, and last elements, pick the median. Better than random in many real cases and avoids sorted-input pathology

    - It's generally faster in practice but trickier to implement correctly.
- different partitioning schemes have different stability properties.

````python
import random
 
# ── helpers ──────────────────────────────────────────────────────────────────
 
def _partition(my_list, start, end):
    """Standard Lomuto partition. Assumes pivot is already at end-1."""
    ptr = start
    pivot = end - 1
    for i in range(start, end):
        if my_list[i] < my_list[pivot]:
            my_list[i], my_list[ptr] = my_list[ptr], my_list[i]
            ptr += 1
    my_list[pivot], my_list[ptr] = my_list[ptr], my_list[pivot]
    return ptr                                                      # returns ptr 
 
````

````python
# ── 1. original (last element as pivot) ──────────────────────────────────────
 
def quick_sort_last(my_list, start=None, end=None):
    if start is None: 
        start = 0
    if end is None:   
        end = len(my_list)
    if end - start <= 1:
        return              # the base case, without this the recursion would never stop
    # pivot is already at end-1, nothing to do
    ptr = _partition(my_list, start, end)
    quick_sort_last(my_list, start=start, end=ptr)
    quick_sort_last(my_list, start=ptr+1, end=end)
````

````python
# ── 2. random pivot ───────────────────────────────────────────────────────────
 
def quick_sort_random(my_list, start=None, end=None):
    if start is None: 
        start = 0
    if end is None:   
        end = len(my_list)
    if end - start <= 1:
        return
    # pick a random index and swap it to end-1
    rand_idx = random.randint(start, end - 1)
    my_list[rand_idx], my_list[end - 1] = my_list[end - 1], my_list[rand_idx]
 
    ptr = _partition(my_list, start, end)
    quick_sort_random(my_list, start=start, end=ptr)
    quick_sort_random(my_list, start=ptr+1, end=end)
````

````python
# ── 3. median-of-three pivot ─────────────────────────────────────────────────
# It looks at three values, figures out which index holds the median, and returns that index.
def _median_of_three(my_list, start, end):
    """Return the index of the median of first, middle, and last elements."""
    mid = (start + end - 1) // 2
    a, b, c = start, mid, end - 1
    # sort the three indices by value and return the middle one
    if my_list[a] > my_list[b]: a, b = b, a
    if my_list[a] > my_list[c]: a, c = c, a
    if my_list[b] > my_list[c]: b, c = c, b
    return b  # b is now the median index
 
# the actual sorting function. It calls _median_of_three to get a good pivot index, then does all the real work — 
# swap pivot into place, partition, recurse.
def quick_sort_median3(my_list, start=None, end=None):
    if start is None: start = 0
    if end is None:   end = len(my_list)
    if end - start <= 1:
        return
    # swap median-of-three to end-1
    med_idx = _median_of_three(my_list, start, end)
    my_list[med_idx], my_list[end - 1] = my_list[end - 1], my_list[med_idx]
 
    ptr = _partition(my_list, start, end)
    quick_sort_median3(my_list, start=start, end=ptr)
    quick_sort_median3(my_list, start=ptr+1, end=end)
````

# Median of Medians explained (as it is confusing)

The simple version is:
- When you find the median of each group of 5, you know at least 2 values in that group are smaller than it and 2 are bigger. So the median is never the worst value in its group.
- When you then find the median of those medians, you're picking something that's "middling" twice over — it survived being middling in its group, then middling again across all the groups.

Ok so we established the median survived being middling twice over. Now let's put numbers on "middling".
- Take a list of 50 elements, split into 10 groups of 5. Each group's median has 2 values below it and 2 above it in that group.
- When you pick the median of the 10 group medians, at least 5 of those medians are below it. Each of those 5 medians has 2 values below them in their group. So:

```python
5 medians below pivot  ×  2 values below each median  =  10 values
10 values + 5 medians themselves                       =  15 values
15 out of 50                                           =  30%
```
- So at least 30% of values are guaranteed below the pivot. By the same logic at least 30% are guaranteed above it. 
- Which means the pivot must land somewhere in the `middle 40%` — between the 30th and 70th percentile.
- With median-of-medians you've mathematically proven the pivot sits somewhere in the middle 40% of values, 
- so the worst split you can ever get is `30/70`. 
- And a 30/70 split repeated all the way down the recursion tree stays `O(n log n)` — it **never blows up into O(n²)**.

## Note

it proves that a guaranteed O(n log n) sort is possible. **But in the real world nobody actually uses it because median-of-three or random pivot are fast enough and much simpler**.

````python
# ── 4. median-of-medians ─────────────────────────────────────────────────
def _partition(my_list, start, end):
    """Standard Lomuto partition. Assumes pivot is already at end-1."""
    ptr = start
    pivot = end - 1
    for i in range(start, end):
        if my_list[i] < my_list[pivot]:
            my_list[i], my_list[ptr] = my_list[ptr], my_list[i]
            ptr += 1
    my_list[pivot], my_list[ptr] = my_list[ptr], my_list[pivot]
    return ptr


def _insertion_sort_small(my_list, start, end):
    """Sort a small sublist in place. Used to find medians of groups of 5."""
    for i in range(start + 1, end):
        key = my_list[i]
        j = i - 1
        while j >= start and my_list[j] > key:
            my_list[j + 1] = my_list[j]
            j -= 1
        my_list[j + 1] = key


def _median_of_medians(my_list, start, end):
    """
    Return the index of a pivot guaranteed to be between the
    30th and 70th percentile of the sublist.

    Steps:
      1. Split sublist into groups of 5
      2. Sort each group, find its median index
      3. Collect all medians into the front of the sublist
      4. Recurse to find the median of those medians
    """
    size = end - start

    # base case — 5 or fewer elements, just sort and return middle
    if size <= 5:
        _insertion_sort_small(my_list, start, end)
        return (start + end - 1) // 2

    # step 1 & 2: sort each group of 5, collect median at front
    median_end = start  # tracks where we're writing medians to
    for group_start in range(start, end, 5):
        group_end = min(group_start + 5, end)
        _insertion_sort_small(my_list, group_start, group_end)
        group_median = (group_start + group_end - 1) // 2
        # swap that group's median to the next available front position
        my_list[median_end], my_list[group_median] = my_list[group_median], my_list[median_end]
        median_end += 1

    # step 3: recursively find the median of those collected medians
    return _median_of_medians(my_list, start, median_end)


def quick_sort_mom(my_list, start=None, end=None):
    if start is None: start = 0
    if end is None:   end = len(my_list)
    if end - start <= 1:
        return

    # get a guaranteed good pivot and swap it to end-1
    med_idx = _median_of_medians(my_list, start, end)
    my_list[med_idx], my_list[end - 1] = my_list[end - 1], my_list[med_idx]

    ptr = _partition(my_list, start, end)
    quick_sort_mom(my_list, start=start, end=ptr)
    quick_sort_mom(my_list, start=ptr+1, end=end)
````

### Hoare Partitioning

- `pivot_val` stores the actual value, not an index — because unlike Lomuto, the pivot doesn't get swapped to a fixed position first, 
- so you'd lose track of it.

- Two pointers `left` and `right` start at opposite ends and move inward. 
- `left` hunts for a `value that's too big (≥ pivot)`, 
- `right` hunts for a `value that's too small (≤ pivot)`, then they swap and keep going until they cross.

- The recursive calls use `right + 1` and `left` as the `split point` rather than `ptr` and `ptr + 1` — 
- this is the **trickiest part of Hoare**. Because the pivot doesn't land in its final position after partitioning, 
- you **split on where the pointers crossed** rather than where the **pivot ended up**.

````python
# ── 5. Hoare partitioning ─────────────────────────────────────────────────
# similar to the reversing a list in place structure we had used earlier in the unit

def quick_sort_hoare(my_list, start=None, end=None):
    if start is None: start = 0
    if end is None:   end = len(my_list)
    if end - start <= 1:
        return

    pivot_val = my_list[(start + end - 1) // 2]        # could adapt to use a random pivot, median-of-three or median of medians like the other cases above

    left = start
    right = end - 1

    while left <= right:
        while my_list[left] < pivot_val:        
            left += 1                                   # left keeps moving until it finds something TOO BIG
        while my_list[right] > pivot_val:
            right -= 1                                  # right keeps moving until it finds something TOO SMALL
        if left <= right:                               # if left <= right check is just making sure the pointers haven't already crossed before doing the swap — if they have crossed it means there's nothing left to fix and you're done.
            # So by the time you reach the if left <= right swap, you know:
            # my_list[left]  is too big  — it should be on the right
            # my_list[right] is too small — it should be on the left  
            my_list[left], my_list[right] = my_list[right], my_list[left]
            left += 1
            right -= 1

    quick_sort_hoare(my_list, start=start, end=right + 1)
    quick_sort_hoare(my_list, start=left, end=end)
````

| Middle element | Picks a value near the true median, partitions stay balanced | **O(n log n)** |
| Median-of-three | Same result, fine here too | **O(n log n)** |
| Median-of-medians | Mathematically guaranteed to pick a pivot between 30th–70th percentile | **O(n log n)** guaranteed |
| Hoare partitioning | Two pointers move inward — sorted input is actually a good case here | **O(n log n)** |

## Reverse sorted `[8, 7, 6, 5, 4, 3, 2, 1]`

| Strategy | What happens | Complexity |
|---|---|---|
| Last element | Pivot is always the min this time, same lopsided problem | **O(n²)** |
| Middle element | Still picks something near the middle, partitions stay balanced | **O(n log n)** |
| Median-of-three | Fine here too | **O(n log n)** |
| Median-of-medians | Guarantee holds regardless of input order | **O(n log n)** guaranteed |
| Hoare partitioning | Handles this well — pointers meet near the middle | **O(n log n)** |

## Pathological partial sort (middle value repeatedly near min or max of each sublist)

#### Pathological in algorithms?

- In algorithms it means an input that's been **specifically constructed to trigger the absolute worst behaviour of an algorithm**. 
- It's not a random bad input — it's a deliberately crafted one that exploits a known weakness.
- So "pathological input for quicksort with last-element pivot" means someone looked at how the algorithm works and 
- engineered a list that guarantees the worst possible pivot choice every single time

| Strategy | What happens | Complexity |
|---|---|---|
| Last element | Lopsided partitions as before | **O(n²)** |
| Middle element | Can hit worst case depending on the pattern | **O(n²)** |
| Median-of-three | Samples three spread-out positions — very unlikely their median is a terrible pivot repeatedly across every recursive call | **O(n log n)** in practice |
| Median-of-medians | Guarantee still holds — no input pattern can break it | **O(n log n)** guaranteed |
| Hoare partitioning | Still vulnerable to bad pivot choice unless combined with median-of-three or random selection | **O(n²)** worst case |

## What is Hoare partitioning?

Hoare is an alternative to Lomuto (what we've been using). Instead of one pointer sweeping left to right, it uses two pointers starting at opposite ends and moving inward until they meet.

```
left pointer  →                ← right pointer
[37, 42, 9, 19, 35, 4, 53, 22]
```

- Left pointer moves right until it finds a value >= pivot
- Right pointer moves left until it finds a value <= pivot
- Swap those two values
- Repeat until the pointers cross

The key differences from Lomuto:

| | Lomuto | Hoare |
|---|---|---|
| Pointers | One pointer, left to right | Two pointers moving inward |
| Pivot position after partition | Pivot lands in its final sorted position | Pivot does NOT end up in its final position |
| Swaps on average | More swaps | Fewer swaps (~3x fewer) |
| Implementation complexity | Simpler | Trickier to get right |
| Worst case complexity | O(n²) with bad pivot | O(n²) with bad pivot |

The important point: Hoare is faster in practice due to fewer swaps, but it doesn't change the worst case complexity on its own. **You still need a good pivot selection strategy (random or median-of-three) on top of it to avoid O(n²)**.

## Key insight

O(n log n) relies on partitions being roughly balanced. **Any pivot strategy that repeatedly produces lopsided splits degrades toward O(n²)**. ]
 
The strategies ranked by reliability:

| Strategy | Worst case | Practical speed |
|---|---|---|
| Last element | O(n²) — common inputs trigger it | Fast when input is random |
| Middle element | O(n²) — less common but possible | Fast |
| Median-of-three | O(n²) — very unlikely in practice | Fast, used in most real implementations |
| Hoare + median-of-three | O(n²) — very unlikely | Faster than Lomuto due to fewer swaps |
| Median-of-medians | O(n log n) — mathematically guaranteed | Slower due to overhead, rarely used in production |

### Merge Sort

Quick sort isn't the only divide and conquer sorting algorithm — meet its close cousin, merge sort. Where `quick sort splits around a pivot and hopes for a balanced divide`, `merge sort always splits exactly down the middle`. 

**The trade-off is that putting the two halves back together — the `merge step` — requires a bit more work.**

#### Here is the idea, using our familiar list:

| 37 | 42 | 9 | 19 | 35 | 4 | 53 | 22 |

**Step 1 — Split** down the middle into two halves:

| 37 | 42 | 9 | 19 |   and   | 35 | 4 | 53 | 22 |

**Step 2 — Recurse**: `merge sort` the left half, `merge sort` the right half. Hopefully this is screaming *recursion* at you again! Trust the `recursive leap of faith` — if merge sort works on smaller lists, we just need to worry about what happens at the bottom and how we put things back together.

The `base case` is a `list of one element` — it's already sorted, **so we just return it.**

**Step 3 — Merge**: take two *sorted* halves and combine them into one sorted list. This is the key insight: merging two already-sorted lists is easy. **Just look at the front of each, pick the smaller one, and repeat.**

So for our example, after recursing all the way down and back up:

| 9 | 19 | 37 | 42 |   merged with   | 4 | 22 | 35 | 53 |

Pick from the fronts: 4 < 9, take 4. Then 9 < 22, take 9. Then 19 < 22, take 19. And so on:

| 4 | 9 | 19 | 22 | 35 | 37 | 42 | 53 |

#### The Merge Step

**The merge step is where all the real work happens**. We compare the **front elements of the two halves**, **take the smaller one**, and **advance that pointer** — repeating until one half is exhausted, then **appending whatever remains** of the other.

Unlike quick sort, **merge sort cannot do this in-place — it needs a temporary list to write results into before copying back**. This is the **key memory cost.**



#### A Note on Stability

Notice that when two equal elements meet during the merge step, we always take the `left` one first. This means **elements that appeared earlier in the original list will still appear earlier in the sorted output** — `merge sort is stable`**` by construction. 

This is in contrast to `quick sort`, where the `pivot swap can disrupt the relative order of equal elements`.

````python
def merge(left, right):
    """Merge two sorted lists into one sorted list."""
    result = []
    i = 0
    j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:   # <= rather than < preserves stability, as you append from the left to result
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    # one half is exhausted — append whatever remains of the other
    result.extend(left[i:])
    result.extend(right[j:])
    return result

# start with your input list
# merge_sort — only cares about splitting. No pointers, just mid
def merge_sort(my_list):
    if len(my_list) <= 1:         # base case: already sorted
        return my_list
    mid = len(my_list) // 2
    left = merge_sort(my_list[:mid])    # recursive leap of faith: trust this returns a sorted left half
    right = merge_sort(my_list[mid:])   # trust this returns a sorted right half
    return merge(left, right)           # now just merge two sorted lists

# Note each call to merge_sort has its own left and right living in memory simultaneously,
# you finish all the merge_sort actions first then trigger merge for all those lists
# Merge sort uses O(n) extra memory — all those intermediate lists exist at the same time. 
# Compare that to quick sort which sorts in-place, so it doesn't need to allocate any of this.


my_list = [37, 42, 9, 19, 35, 4, 53, 22]
print(merge_sort(my_list))
````

### Merge Sort Complexity

***Try to work out the complexity of merge sort before continuing.***

The structure mirrors quick sort's reasoning almost exactly. The merge step scans through all $n$ elements to produce the combined output — that's $O(n)$ work. And because we always split exactly in half, the depth of the recursion is always $\log(n)$, with no worst-case risk of lopsided splits.

This gives `merge sort` a complexity of $O(n \log n)$ — and crucially, this holds in **all cases**: best, average, *and* worst. 
`Quick sort` can degrade to $O(n^2)$ on sorted or adversarial input; merge sort never does.

The cost of this guarantee is **memory**. Unlike quick sort, which sorts in-place (using only $O(\log n)$ memory for the call stack), `merge sort` **creates new lists at every merge step**. This means it uses $O(n)$ additional memory overall — a **meaningful difference for very large datasets.**

| | Quick Sort | Merge Sort |
|---|---|---|
| Best case | $O(n \log n)$ | $O(n \log n)$ |
| Average case | $O(n \log n)$ | $O(n \log n)$ |
| Worst case | $O(n^2)$ | $O(n \log n)$ |
| Memory | $O(\log n)$ | $O(n)$ |
| Stable? | No | Yes |
| In-place? | Yes | No |

In practice, quick sort is usually faster despite the same average complexity — the constant factors are lower because it avoids allocating new lists. But when you need a **guaranteed** $O(n \log n)$ or a **stable** sort, merge sort is the go-to. 

Python's own built-in `sorted()` function actually uses a variant called *Timsort*, which **combines merge sort with insertion sort** for small sublists.

```python
quick_sort(my_list, start=ptr+1, end=end)
```
### Solution:
- You could do the same with merge sort — `pass start, mid, end and avoid slicing`. That way there's only ever one list in memory.
-  yes, you can reduce the number of intermediate lists by passing indices, but you can't get merge sort down to O(1) extra memory the way quick sort manages. The merge step always needs breathing room. The best you can do is reuse a single temporary buffer across all the merge calls rather than allocating fresh ones every time.

### Trade-off
- Quick sort could work in-place because swapping elements around the pivot doesn't need extra space. 
- But merge fundamentally can't do its job without somewhere to write the combined result — you can't merge two halves of the same list back into itself without overwriting values you still need to read.
- So even if you pass indices and avoid the slicing overhead, you still need a temporary buffer for the merge step. You've saved on the recursive splitting cost, but you can't escape needing some extra memory for merging.
