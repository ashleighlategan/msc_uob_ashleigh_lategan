## 6.2 – Sorting
### Selection Sort
Hopefully you're happy with the idea of sorting from the material on Engage, so let's cut to the chase and introduce the first sorting algorithm: *selection sort*.

The basic idea behind selection sort is to build up the sorted list of items by repeatedly taking the smallest item from the input list and moving it into the correct place in the sorted list.

#### Implementation
As we move towards pseudocode and eventually the implementation itself we must make more and more details concrete. Below is an implementation of selection sort in Python – notice it is a procedure, it modifies the list in-place.

As an exercise, try writing down a set of steps as pseudocode – code-like syntax that is not specific to Python. Did the choice of Python force us to make any specific decisions we might not have if we were explaining it in a more abstract form?

````python
def selection_sort(my_list):
    for i in range(0, len(my_list)-1):
        # look for smallest item in rest of list, to swap with item in position i
        
        min_index = i
        for j in range(i + 1, len(my_list)):
            if my_list[j] < my_list[min_index]:
                min_index = j
                
        if i != min_index:
            my_list[i], my_list[min_index] = my_list[min_index], my_list[i]
            
my_list = [37, 42, 9, 19, 35, 4, 53, 22]
selection_sort(my_list)
print(my_list)
````

## code explanation

- i marks the "boundary" position — the slot you're trying to fill with the correct (minimum) value. It also doubles as your initial guess for where the minimum is (min_index = i).
- j is the "scout" — it walks through everything to the right of i, looking for something smaller than your current best guess. Every time it finds one, it updates min_index.

### helpful notes:

- min_index = i at the start of each outer loop iteration, so when i=1, min_index starts as 1, not 0.
    - The reason is that index 0 is already sorted and locked. The algorithm only cares about the unsorted portion to the right of i.

## Simplified explanation

#### The big picture ideaThe big picture idea

- Imagine playing cards are face down on a table in a random row. 
- You scan all of them, flip them up one by one to find the smallest, then pick it up and place it at the far left. 
- Now that position is done forever — you never touch it again. 
- Then you scan the remaining cards for the next smallest, place it in position 2. And so on.
- That's selection sort. You're repeatedly selecting the minimum from the unsorted region and slotting it into place.

```python 
[sorted region | unsorted region]
```



#### What i is

- `i` is the position you're currently filling — the next slot in the sorted region.
```python 
for i in range(0, len(my_list)-1):

i=0: [? | 37, 42, 9, 19, 35, 4, 53, 22]   ← find smallest, put it at position 0
i=1: [4 | ? , 42, 9, 19, 35, 37, 53, 22]   ← find smallest, put it at position 1
i=2: [4, 9 | ?, 42, 19, 35, 37, 53, 22]    ← and so on
```
- We stop at `len(my_list)-1` because by the time the last slot is reached, it must already be the largest — **there's nothing left to compare against**.

#### What j and min_index are
- `j` scans everything to the right of ``i` looking for the smallest item. 
- `min_index` tracks where that smallest item lives.

```python
min_index = i  # assume position i is the smallest until proven otherwise

for j in range(i + 1, len(my_list)):
    if key(my_list[j]) < key(my_list[min_index]):
        min_index = j  # found something smaller, update
```
- At the **end of the inner loop***, `min_index` points to the **smallest item** in the **unsorted** region.

#### The swap
- Put the **smallest item** into position `i`, and **move** whatever was at `i` into the **gap left behind**.
- The if `i != min_index` check just **avoids a pointless swap** when the smallest item is already in the right place.

```python
if i != min_index:
    my_list[i], my_list[min_index] = my_list[min_index], my_list[i]
```

#### Tracing i=0

```python
Starting list: [37, 42, 9, 19, 35, 4, 53, 22]
min_index = 0  (value 37)
j=1: 42 < 37? No
j=2: 9  < 37? Yes → min_index = 2
j=3: 19 < 9?  No
j=4: 35 < 9?  No
j=5: 4  < 9?  Yes → min_index = 5
j=6: 53 < 4?  No
j=7: 22 < 4?  No

min_index = 5, swap positions 0 and 5
→ [4, 42, 9, 19, 35, 37, 53, 22]
```
- **Position 0 is now fixed forever.**

#### The key parameter

```python
def selection_sort(my_list, key=lambda x:x):
```
- `key` is a **function applied to each item before comparing**. The default `lambda x: x` just returns the item itself, so by default it sorts by value directly.
- But you could sort your PlayingCard objects by number without having __lt__ defined:
- This is the same idea as Python's built-in `sorted(my_list, key=...)` 
    - it separates how to compare from what to sort, making the function flexible for any type of object.
```python
selection_sort(cards, key=lambda card: card.number)
```


#### Difference between selection and insertion sort:
- In `insertion sort` you pick up the next card in sequence and then **scan backwards to find its correct position** among **already sorted cards**:
    - Take card at position k → scan left to find where it fits
    - Insertion sort does small scans (only the sorted region) but potentially many shifts to make room
- In `selection sort` you **scan the entire unsorted region** to find the **minimum**, then place it:
    - Scan entire unsorted region → take the minimum and place it at position i
    - Selection sort does one big scan of the unsorted region but only one swap per iteration

#### Insertion sort at k=2 (inserting 9):
```python
temp = 9
scan left: 42 > 9 → shift right
scan left: 37 > 9 → shift right
hit start  → place 9 at position 0
```
#### Selection sort at i=0:
```python
scan entire unsorted region: 37, 42, 9
minimum is 9 at index 2
swap positions 0 and 2 → done in one move
```


#### Selection Sort Complexity
When considering the time complexity of sorting algorithms, we usually consider the *comparison* to be the constant time element, as with search. This makes selection sort quite easy to analyse since it always performs the same number. On the first iteration for an input of size $n$ it will check $n-1$ items to find the smallest, on the next iteration it will check $n-2$, then $n-3$ and so on. 

The exact sum is quite easy to calculate using the formula for the sum of consecutive integers, but remember we are not *really* interested in the exact sum. The biggest term in the final formula will be $n^2$, meaning we have a complexity of $O(n^2)$, we do not care about any other terms in the sum or constant factors. It's actually quite easy to see this complexity arise directly from the code: there are two nested for loops, both of which do approximately $n$ iterations, so the final result should be $n \times n = n^2$.

In addition, the complexity is always the same: selection sort is best, worst, and average case complexity of $O(n^2)$.

## Sum Formulas

### Consecutive integers (summing values)
$$1 + 2 + 3 + \dots + n = \frac{n(n+1)}{2}$$

### Selection sort (summing comparisons per iteration)
For a list of size $n$, each pass does one fewer comparison:

$$(n-1) + (n-2) + (n-3) + \dots + 1 = \frac{n(n-1)}{2}$$

### Why n-1 comparisons on the first pass?
For $n = 4$:

| Iteration | Comparisons | Formula |
|-----------|-------------|---------|
| $i=0$     | 3           | $n-1$   |
| $i=1$     | 2           | $n-2$   |
| $i=2$     | 1           | $n-3$   |
| **Total** | **6**       | $\frac{4(4-1)}{2} = \frac{12}{2} = 6$ |

### Big-O
$$\frac{n(n-1)}{2} = \frac{n^2 - n}{2} \rightarrow O(n^2)$$

#### Instability
The final thing to point out is that selection sort is an *unstable* sorting algorithm in this implementation. This means that equal items are not guaranteed to be in the same order after sorting. 

That might initially seem like an odd idea: surely `9` always equals `9`? Well, it does! If you are sorting a list of integers you have nothing to worry about.

But imagine you were sorting cards by number: the 9 of spades might be before the 9 of hearts in the input, but *after* in the output. 

To demonstrate, we need to tweak the function. If we compare Python tuples using `<` it will compare each item in turn:

````python
(9, "♥") < (9, "♠")
````

````python
(9, "♥") > (9, "♠")
````

The strings `"♥"` and `"♠"` are automatically compared to break the tie between the two 9s. If we sort a list of tuples that represent playing cards, it will **sort by number first, then by whichever suit happens to have the lower character encoding**. We don't want that – a stable short should be able to sort by the first element and leave all the other elements in their original order.

Let's enhance our selection sort function with a `key` parameter, which takes a function to be applied to each element in the list being sorted. By default it will still compare each item whole, but it will allow us to customise the behaviour from the perspective of the caller.

````python
def selection_sort(my_list, key=lambda x:x):
    for i in range(0, len(my_list)-1):
        # look for smallest item in rest of list, to swap with item in position i
        
        min_index = i
        for j in range(i + 1, len(my_list)):
            if key(my_list[j]) < key(my_list[min_index]):
                min_index = j
                
        if i != min_index:
            my_list[i], my_list[min_index] = my_list[min_index], my_list[i]
            
my_list = [37, 42, 9, 19, 35, 4, 53, 22]
selection_sort(my_list)
print(my_list)
````

Now the key is applied to each item being compared: take a look at the code change and make sure you understand it!

You can see the old example still works fine, also.

Now we can call the function and provide a new `key` function which says to only consider the first item when looking for the minimum:

````python
my_cards = [(9, "♠"), (9, "♥"), (4, "♣")]
selection_sort(my_cards, key=lambda t: t[0])
print(my_cards)
````

### Note:
The only reason it looks weird with two brackets is because normally functions have names:
```python
get_number(my_list[j])   # named function - looks normal
(lambda t: t[0])(my_list[j])  # anonymous function - looks odd
```

## Selection Sort with Key - Walking Through `my_cards`

**Starting list:** `[(9, "♠"), (9, "♥"), (4, "♣")]`  
**key:** `lambda t: t[0]` (extract the number only, ignore suit)

---

### i=0, min_index=0
Guessing `(9, "♠")` is min → key gives us `9`

| j | Comparison | Result | min_index |
|---|-----------|--------|-----------|
| 1 | `key((9, "♥")) < key((9, "♠"))` → $9 < 9$? | No | 0 |
| 2 | `key((4, "♣")) < key((9, "♠"))` → $4 < 9$? | Yes | 2 |

Swap index 0 and 2 → `[(4, "♣"), (9, "♥"), (9, "♠")]`

---

### i=1, min_index=1
Guessing `(9, "♥")` is min → key gives us `9`

| j | Comparison | Result | min_index |
|---|-----------|--------|-----------|
| 2 | `key((9, "♠")) < key((9, "♥"))` → $9 < 9$? | No | 1 |

No swap → `[(4, "♣"), (9, "♥"), (9, "♠")]`

---

### Final result: `[(4, "♣"), (9, "♥"), (9, "♠")]`

The two 9s **stayed in their original order** because $9 < 9$ was never `True`,
so the suits were never compared at all.

If you follow the logic of the algorithm, I hope it is pretty easy to see why, despite the fact that we made sure the function only considered the first item, we still ended up with the order of the two 9-cards reversed compared to our original list. 

Notice as we described above, the default ordering of the suit strings would put `"♠"` before `"♥"` *anyway*, so at least the `key` argument is definitely working!
