#### Bubble Sort
Bubble sort works by iterating over the list, comparing adjacent items, and swapping them if they are out of order. 

Notice that after the first pass of a bubble sort, the biggest item is always moved to the end of the list. This means the next iteration does not need to check the final position of the list. 

In addition, you can keep track of how many items were swapped on each iteration. If this count is zero on any iteration, then the list must be fully sorted, in which case the algorithm can stop early.

#### Merge Sort
To understand merge sort, let's first understand the *merge* operation. The merge operation takes two *sorted* lists and combines them into another sorted list. Imagine we are merging `list1` and `list2`, both of which are sorted. Create two variables, one pointing to the first element (zero) of each list, call this `ptr1` and `ptr2`.

Then, go through each index of the merged list (e.g. `i in range(0, len(list1)+len(list2))`). Set position `i` of the merged list to the *smaller* value of `list1[ptr1]` and `list2[ptr2]`, then advance the corresponding pointer by one.

So initially you will compare the first element of each list, if the value in `list1` is smaller then copy this into the merged list, and then compare the second item of `list1` with the first item of `list2`.

Keep going until one pointer goes beyond its list – at this point, copy the rest of the other list into the merged list.

Now, to understand *merge sort*, realise that any list of size 1 must be considered sorted. So, here is the process for the input list:
* If `in_list` is length 1, return `in_list`
* Otherwise, split `in_list` list into two halves: `left` and `right`
* Merge sort `left`
* Merge sort `right`
* Merge `left` and `right` into `merged`
* Return `merged`

For any size input list, this will be recursively split into two until both halves are length 1, and then the final sorted list will be rebuilt by merging sorted lists.

```python
def bubble_sort(in_list):
    new_list = in_list.copy()
    total = len(new_list)
    for i in range(total):
        swap = False
        for j in range(0, total-i-1):           # -i sine the last i elements are already sorted
            if new_list[j] > new_list[j+1]:
                new_list[j], new_list[j+1] = new_list[j+1], new_list[j]
                swap = True
        if not swap:
            break                               # if no swaps on this run then the list is already sorted and we can stop 
    return new_list
    
def merge_sort(in_list):
    if len(in_list) <=1:
        return in_list                          # Base case where a list 1 elements is already sorted and list of 0 elements doesn't need sorting
    
    middle = len(in_list) // 2
    left = merge_sort(in_list[:middle])         # Recursively sorting the left half of the list
    right = merge_sort(in_list[middle:])        # Recursively sorting the right half of the list
    return merge(left, right)

def merge(left, right):
    merged_list = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged_list.append(left[i])
            i += 1
        else:
            merged_list.append(right[j])
            j += 1
    # Append any of the remaining items from either list
    merged_list += left[i:]
    merged_list += right[j:]
    return merged_list
```

#### Time complexity
Time complexity for the **worst case** is `O(n^2)` as you can't just loop over every element of the list once as that would just push the largest item to the right, you need to go through several iterations to move each larger item to the right; this could be up to `total` moves where total is `n` so because in the worst case inner loop and outer loop would review each item up `n items` and have to swap on each pass, your worst case is O(n^2). 

The **best case** is `O(n)` if the list is **already sorted** as then no swaps are needed, and only one pass of the inner loop takes place and due to no swaps triggering the break and return. 

The **average case** I found harder to work out as initially I thought that the break optimization may help make this better than `O(n^2)` but that break only triggers when the list is already sorted on a pass. I then **generalized this** since **each item** of `n` could be **equally likely** to be anywhere, the **average distances** = `(0 + 1 + 2 + 3.....(n-1))/n` and summing that I used the trick from **Gauss/Triangular numbers** that becomes `n(n-1)/2/n` = `(n-1)/2` ~ `n/2`. Then if each element is n/2 places away from its final position and each element can move one step per pass then we would need n/2 passes before elements are able to trigger the break so that is `~ O(n)` passes since with Big O we can drop constants as this shows it still scales with n. Each pass also has an inner loop where we range from `j = 0` to `j = total - i-1` where `total` is `n` so the number of passes shrinks as we go but is still **proportional** to `n` as we have `n` within the **upper bound** so then comparisons per pass would average to  ~ `n/2` so `~ O(n)`.
